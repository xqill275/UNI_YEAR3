### Removing duplicates

```python
# Create a DataFrame object
DF_obj= DataFrame({'column 1':[1,1,2,2,3,3,3],
                   'column 2':['a', 'a','b', 'b', 'c', 'c', 'c'],
                   'column 3':['A', 'A', 'B', 'B', 'C', 'C', 'C']})
DF_obj
```

```python
# Check duplicates in rows
# Run the following and discuss the output
DF_obj.duplicated()
```

```python
# Drop duplicates
DF_obj.drop_duplicates()
```
