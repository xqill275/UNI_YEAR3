# Data Preparation Basics
## Part 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = pd.DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3))
DF_obj_2
```

### Concatenating data

```python
# Concatenate horizontally (side by side)
pd.concat([DF_obj, DF_obj_2], axis=1)
```

```python
# Concatenate vertically
pd.concat([DF_obj, DF_obj_2])
```

```python
# Concatenate two equal dataframes
# Sample DataFrames
DF_obj = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
DF_obj_2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]})

# Concatenate horizontally (side by side)
result = pd.concat([DF_obj, DF_obj_2], axis=1)

result

# Run the code and check did we get a decimal point as above and why?
```

### Transforming data
#### Dropping data

```python
# Remove specific rows from a DataFram
DF_obj.drop([0, 2])
```

```python
# Remove specific columns from a DataFram
DF_obj.drop([0, 2], axis=1)
```

### Adding data

```python
series_obj = Series(np.arange(6))
series_obj.name = "added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF_obj, series_obj)
variable_added
```

### Sorting data

```python
DF_sorted = DF_obj.sort_values(by=(5), ascending=[False])
DF_sorted
```
